# Descriptive Statistics

**Objective**: *Use descriptive statistics to gather information about a data set.*

In this lab, you will complete a series of exercises to calculate various summary statistics of the `dsfda.ht_daily_metrics` data set.

```
%run "../../Includes/Classroom-Setup"
```

## Part 1: Measures of Central Tendency

-sandbox
#### Exercise 1: Mean

Use the `AVG` SQL function to calculate the average resting heartrate, as an integer, for each lifestyle.

<img alt="Hint" title="Hint" style="vertical-align: text-bottom; position: relative; height:1.75em; top:0.3em" src="https://files.training.databricks.com/static/images/icon-light-bulb.svg"/>&nbsp;**Hint:** If needed, check out the [Spark SQL documentation on `avg`](https://spark.apache.org/docs/3.0.1/api/sql/index.html#avg).

```
%sql
-- ANSWER

SELECT CAST(AVG(resting_heartrate) AS int) as avg_resting_heartrate, lifestyle
FROM dsfda.ht_daily_metrics
GROUP BY lifestyle
```

-sandbox
#### Exercise 2: Median

Use the `percentile` SQL function to calculate the median resting heartrate, as an integer, for the "Athlete" lifestyle.

Note that this code block takes longer to run; this is because calculating the median requires shuffling across worker nodes.

<img alt="Hint" title="Hint" style="vertical-align: text-bottom; position: relative; height:1.75em; top:0.3em" src="https://files.training.databricks.com/static/images/icon-light-bulb.svg"/>&nbsp;**Hint:** If needed, check out the [Spark SQL documentation on `percentile`](https://spark.apache.org/docs/3.0.1/api/sql/index.html#percentile).

```
%sql
-- ANSWER

SELECT CAST(percentile(resting_heartrate, 0.5) AS int) AS median_resting_heartrate
FROM dsfda.ht_daily_metrics
WHERE lifestyle = 'Athlete'
```

-sandbox
#### Exercise 3: Mode

Calculate the mode, as an integer, of the resting heartrate for each lifestyle.

<img alt="Hint" title="Hint" style="vertical-align: text-bottom; position: relative; height:1.75em; top:0.3em" src="https://files.training.databricks.com/static/images/icon-light-bulb.svg"/>&nbsp;**Hint:** Cast the heartrate to the nearest integer first. You can use the Spark SQL documentation on `cast` for reference.

<img alt="Hint" title="Hint" style="vertical-align: text-bottom; position: relative; height:1.75em; top:0.3em" src="https://files.training.databricks.com/static/images/icon-light-bulb.svg"/>&nbsp;**Hint:** Try to find the `mode` Spark SQL function in the documentation on your own.

```
%sql
-- ANSWER

SELECT COUNT(*) AS count, CAST(resting_heartrate AS int) AS resting_heartrate
FROM dsfda.ht_daily_metrics
GROUP BY resting_heartrate
ORDER BY count DESC
```

## Part 2: Measures of Dispersion

-sandbox
#### Exercise 4: Standard Deviation

Calculate the standard deviation of the resting heartrate for each lifestyle.

<img alt="Hint" title="Hint" style="vertical-align: text-bottom; position: relative; height:1.75em; top:0.3em" src="https://files.training.databricks.com/static/images/icon-light-bulb.svg"/>&nbsp;**Hint:** Cast the heartrate to the nearest integer first.

<img alt="Hint" title="Hint" style="vertical-align: text-bottom; position: relative; height:1.75em; top:0.3em" src="https://files.training.databricks.com/static/images/icon-light-bulb.svg"/>&nbsp;**Hint:** Use the Spark SQL documentation to determine which function computes the standard deviation.

```
%sql
-- ANSWER

SELECT CAST(STD(resting_heartrate) AS int) as std_resting_heartrate, lifestyle
FROM dsfda.ht_daily_metrics
GROUP BY lifestyle
```

-sandbox
#### Exercise 5: Interquartile Range

Calculate the interquartile range of resting heartrate, as an integer, for each lifestyle.

<img alt="Hint" title="Hint" style="vertical-align: text-bottom; position: relative; height:1.75em; top:0.3em" src="https://files.training.databricks.com/static/images/icon-light-bulb.svg"/>&nbsp;**Hint:** Recall that the interquartile range is the difference between the 75th percentile and the 25th percentile.

<img alt="Hint" title="Hint" style="vertical-align: text-bottom; position: relative; height:1.75em; top:0.3em" src="https://files.training.databricks.com/static/images/icon-light-bulb.svg"/>&nbsp;**Hint:** Refer to previous exercises in this lab to determine which Spark SQL function can be used to compute the value at a given percentile.

```
%sql
-- ANSWER

SELECT CAST(percentile(resting_heartrate, 0.75) - percentile(resting_heartrate, 0.25) AS int) AS iqr_resting_heartrate
FROM dsfda.ht_daily_metrics
GROUP BY lifestyle
```
